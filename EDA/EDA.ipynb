```python
# Importing relevant libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns

pd.set_option("display.max_columns", 250)
pd.set_option("display.max_rows", 250)
```

```python
# Reading in the dataset

path = "diabetes.csv"

diabetes = pd.read_csv(path)
```

```python
diabetes.head()
```

```python
diabetes.info()
```

```python
# Creating two datasets corresponding to has diabetes and does not have diabetes in order to explore better

negative = diabetes[diabetes.Outcome == 0]
postive = diabetes[diabetes.Outcome == 1]
```

```python
postive.describe()
```

```python
negative.describe()
```

```python
# Here we are creating a correlation matrix in order to see the relationships between each variable

corrMatrix = diabetes.corr()
fig, ax = plt.subplots(1,1, figsize=(15,10))
ax = sns.heatmap(corrMatrix, annot=True, cmap="Blues")
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
plt.show()
```

```python
# Here, we are plotting the histograms for all variables in order to understand distribution better

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(20, 15))
plt.style.use("seaborn")

sns.distplot(diabetes["Pregnancies"], ax=ax1, color="#3EACEF")
ax1.set_title("The Distribution of Pregnancies", fontsize=20)
ax1.set_xlabel("Pregnancies", fontsize=15)
ax1.tick_params(labelsize=15)

sns.distplot(diabetes["Glucose"], ax=ax2, color="#3EACEF")
ax2.set_title("The Distribution of Glucose", fontsize=20)
ax2.set_xlabel("Glucose", fontsize=15)
ax2.tick_params(labelsize=15)

sns.distplot(diabetes["BloodPressure"], ax=ax3, color="#3EACEF")
ax3.set_title("The Distribution of Blood Pressure", fontsize=20)
ax3.set_xlabel("Blood Pressure", fontsize=15)
ax3.tick_params(labelsize=15)

sns.distplot(diabetes["SkinThickness"], ax=ax4, color="#3EACEF")
ax4.set_title("The Distribution of Skin Thickness", fontsize=20)
ax4.set_xlabel("Skin Thickness", fontsize=15)
ax4.tick_params(labelsize=15);
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(20, 15))
plt.style.use("seaborn")

sns.distplot(diabetes["Insulin"], ax=ax1, color="#3EACEF")
ax1.set_title("The Distribution of Insulin", fontsize=20)
ax1.set_xlabel("Insulin", fontsize=15)
ax1.tick_params(labelsize=15)

sns.distplot(diabetes["BMI"], ax=ax2, color="#3EACEF")
ax2.set_title("The Distribution of BMI", fontsize=20)
ax2.set_xlabel("BMI", fontsize=15)
ax2.tick_params(labelsize=15)

sns.distplot(diabetes["DiabetesPedigreeFunction"], ax=ax3, color="#3EACEF")
ax3.set_title("The Distribution of Diabetes Pedigree Function", fontsize=20)
ax3.set_xlabel("Blood Pressure", fontsize=15)
ax3.tick_params(labelsize=15)

sns.distplot(diabetes["Age"], ax=ax4, color="#3EACEF")
ax4.set_title("The Distribution of Age", fontsize=20)
ax4.set_xlabel("Age", fontsize=15)
ax4.tick_params(labelsize=15);
```

```python
# Here, we are plotting some variables against each other to see whether there are any more interesting relationships

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(20, 15))

sns.lineplot(x="Age", y="Pregnancies", color="#3EACEF",
             data=diabetes, ax=ax1)
ax1.set_title("Pregnancies given Age", fontsize=20)
ax1.set_xlabel("Age", fontsize=15)
ax1.set_ylabel("Pregnancies", fontsize=15)
ax1.tick_params(labelsize=15)

sns.lineplot(x="BloodPressure", y="Insulin", hue="Outcome", palette="Blues",
             data=diabetes, ax=ax2)
ax2.set_title("Blood Pressure Levels given Insulin Levels", fontsize=20)
ax2.set_xlabel("Blood Pressure", fontsize=15)
ax2.set_ylabel("Insulin", fontsize=15)
ax2.tick_params(labelsize=15)

sns.scatterplot(x="Glucose", y="Insulin", hue="Outcome", palette="Blues",alpha=0.7,
                data=diabetes, ax=ax3)
ax3.set_title("Glucose given Insulin", fontsize=20)
ax3.set_xlabel("Glucose", fontsize=15)
ax3.set_ylabel("Insulin", fontsize=15)
ax3.tick_params(labelsize=15)

sns.boxplot(x="Outcome", y="Age", color="#3EACEF",data=diabetes, ax=ax4)
ax4.set_title("Box Plot of Age and Outcome", fontsize=20)
ax4.set_xlabel("Outcome", fontsize=15)
ax4.set_ylabel("Age", fontsize=15)
ax4.tick_params(labelsize=15)
```

```python
# Lastly, we plot violiplots of key variables against the target variable of Outcome - here we find very intersting
# relationships with respect to Outcome

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(20, 15))

sns.violinplot(x="Outcome", y="Pregnancies", color="#3EACEF",
             data=diabetes, ax=ax1)
ax1.set_title("Outcome given Pregnancies", fontsize=20)
ax1.set_xlabel("Outcome", fontsize=15)
ax1.set_ylabel("Pregnancies", fontsize=15)
ax1.tick_params(labelsize=15)

sns.violinplot(x="Outcome", y="Glucose", hue="Outcome", palette="Blues",
             data=diabetes, ax=ax2)
ax2.set_title("Outcome given Glucose", fontsize=20)
ax2.set_xlabel("Outcome", fontsize=15)
ax2.set_ylabel("Glucose", fontsize=15)
ax2.tick_params(labelsize=15)

sns.violinplot(x="Outcome", y="Insulin", hue="Outcome", palette="Blues",alpha=0.7,
                data=diabetes, ax=ax3)
ax3.set_title("Outcome given Insulin", fontsize=20)
ax3.set_xlabel("Outcome", fontsize=15)
ax3.set_ylabel("Insulin", fontsize=15)
ax3.tick_params(labelsize=15)

sns.violinplot(x="Outcome", y="Age", color="#3EACEF",data=diabetes, ax=ax4)
ax4.set_title("Outcome and Age", fontsize=20)
ax4.set_xlabel("Outcome", fontsize=15)
ax4.set_ylabel("Age", fontsize=15)
ax4.tick_params(labelsize=15)
```
